```python
# macbook에서 동작하지 않습니다.

from gensim.models import Word2Vec
from gensim.utils import simple_preprocess

# 1. 학습용 데이터 준비 (간단한 문장들)
sentences = [
    "I love machine learning and artificial intelligence",
    "Deep learning is a subset of machine learning",
    "Word embeddings are vector representations of words",
    "Natural language processing is fun",
    "Large language models are powerful"
]

# 2. 전처리: 소문자화 및 토큰화
tokenized_sentences = [simple_preprocess(sentence) for sentence in sentences]

# 3. Word2Vec 모델 학습
# vector_size: 임베딩 차원 (보통 100~300)
# window: 주변 단어를 몇 개까지 볼 것인지
# min_count: 최소 등장 횟수 미만 단어는 무시
model = Word2Vec(sentences=tokenized_sentences, vector_size=10, window=3, min_count=1, sg=1)

# 4. 결과 확인
word = "learning"
vector = model.wv[word]

print(f"'{word}'의 임베딩 벡터 (10차원): \n{vector}\n")

# 5. 유사한 단어 찾기
similar_words = model.wv.most_similar("learning", topn=3)
print(f"'{word}'와 가장 유사한 단어들: {similar_words}")
```
